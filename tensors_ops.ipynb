```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
#Reshape and view
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
#reshape if we dont know then number of items using -1
my_torch2 = torch.arange(10)
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
#View
my_torch3 = my_torch.view(2,5)
my_torch3
```

Basically reshape and view are the same
