```python
import torch
```

```python
tensor_a = torch.tensor([1,2,3,4,5])
tensor_b = torch.tensor([6,7,8,9,10])
```

```python
#Addition
tensor_a + tensor_b
```

```python
#Add function
torch.add(tensor_a, tensor_b)
```

```python
#Subtraction
tensor_a - tensor_b
```

```python
#Sub function
torch.sub(tensor_a, tensor_b)
```

```python
#Multiplication
tensor_a * tensor_b
```

```python
#Mul function
torch.mul(tensor_a, tensor_b)
```

```python
#division
tensor_a / tensor_b
```

```python
#Div function
torch.div(tensor_a, tensor_b)
```

```python
#Remainder Modulus
tensor_b % tensor_a
```

```python
#remainder function
torch.remainder(tensor_b, tensor_a)
```

```python
#Exponents and power
torch.pow(tensor_a, tensor_b)
```

```python
#Another way to write long hand
tensor_a.add(tensor_b)
```

```python
#Reassignment
tensor_c = tensor_a.add_(tensor_b)
tensor_c
```

```python
#or
tensor_a += tensor_b
tensor_a
```

```python
#or
tensor_a = tensor_a + tensor_b
tensor_a
```
